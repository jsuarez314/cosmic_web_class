```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
model_1 = np.loadtxt('./data/envs_bsk+rf.txt')
model_1_class, model_1_counts = np.unique(model_1, return_counts=True)
```

```python
if not(3 in model_1_class):
    model_1_class = np.append(model_1_class,3)
    model_1_counts = np.append(model_1_counts,0)
```

```python
model_21 = np.loadtxt('./data/envs_bsk+fakepoints_1.txt')
model_21_class, model_21_counts = np.unique(model_21, return_counts=True)

model_22 = np.loadtxt('./data/envs_bsk+fakepoints_2.txt')
model_22_class, model_22_counts = np.unique(model_22, return_counts=True)

model_31 = np.loadtxt('./data/envs_voronoi+fakepoints_1.txt')
model_31_class, model_31_counts = np.unique(model_31, return_counts=True)

model_32 = np.loadtxt('./data/envs_voronoi+fakepoints_2.txt')
model_32_class, model_32_counts = np.unique(model_32, return_counts=True)
```

```python
model_1_class
```

```python
X = ['Model 1','Model 2.1','Model 2.2','Model 3.1','Model 3.2']
X_axis = np.arange(len(X))

void_pop = [model_1_counts[3], model_21_counts[3], model_22_counts[3], model_31_counts[3], model_32_counts[3]]
sheet_pop = [model_1_counts[2], model_21_counts[2], model_22_counts[2], model_31_counts[2], model_32_counts[2]]
fila_pop = [model_1_counts[1], model_21_counts[1], model_22_counts[1], model_31_counts[1], model_32_counts[1]]
peak_pop = [model_1_counts[0], model_21_counts[0], model_22_counts[0], model_31_counts[0], model_32_counts[0]]

fig = plt.figure(figsize=(8,4))
plt.bar(X_axis, void_pop, 0.1)
plt.xticks(X_axis, X)
plt.xlabel("Voids")
plt.ylabel("Population")
plt.show()

fig = plt.figure(figsize=(8,4))
plt.bar(X_axis, sheet_pop, 0.1)
plt.xticks(X_axis, X)
plt.xlabel("Sheets")
plt.ylabel("Population")
plt.show()

fig = plt.figure(figsize=(8,4))
plt.bar(X_axis, fila_pop, 0.1)
plt.xticks(X_axis, X)
plt.xlabel("Filaments")
plt.ylabel("Population")
plt.show()

fig = plt.figure(figsize=(8,4))
plt.bar(X_axis, peak_pop, 0.1)
plt.xticks(X_axis, X)
plt.xlabel("Peaks")
plt.ylabel("Population")
plt.show()
```
