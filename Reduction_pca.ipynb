```python
import pandas as pd
import numpy as np
import pylab as pl

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
tng_photo  = pd.read_csv('./data/tng_features.csv').drop(['Unnamed: 0'], axis=1)
sdss_photo = pd.read_csv('./data/stellar_sdss_phot.csv').drop(['Unnamed: 0'], axis=1)
tweb_df = pd.read_csv('./data/tweb_cw_grid256_rcellsize.csv').drop(['Unnamed: 0'], axis=1)
```

```python
df = pd.merge(tng_photo, tweb_df, on='ID')
df = pd.merge(sdss_photo, df, on='ID')
df
```

```python
predictors = ['tng_MU', 'tng_MB', 'tng_MV', 'tng_MK', 'tng_Mg', 'tng_Mr', 'tng_Mi', 'tng_Mz']
data = df[predictors]
data = StandardScaler().fit_transform(data)

model = PCA(n_components=2)
projection = model.fit_transform(data)

pl.scatter(projection[:,0][df['Environment']==0], projection[:,1][df['Environment']==0], label='Void', color='k', s=0.2)
pl.scatter(projection[:,0][df['Environment']==1], projection[:,1][df['Environment']==1], label='Sheet', color='g', s=0.2)
pl.scatter(projection[:,0][df['Environment']==2], projection[:,1][df['Environment']==2], label='Filament', color='orange', s=0.2)
pl.scatter(projection[:,0][df['Environment']==3], projection[:,1][df['Environment']==3], label='Peak', color='r', s=0.2)
pl.legend()
pl.xlim(-5, 15)
pl.ylim(-1,1.5)
pl.show()

# fig = pl.figure(figsize=(10,10))
# pl.subplot(1,2,1)
# pl.scatter(projection[:,0][df['galaxy_class']==0], projection[:,1][df['galaxy_class']==0], label='Blue', color='b', s=0.2)
# pl.xlim(-5, 15)
# pl.ylim(-1,1.5)
# pl.subplot(1,2,2)
# pl.scatter(projection[:,0][df['galaxy_class']==1], projection[:,1][df['galaxy_class']==1], label='Red', color='r', s=0.2)
# pl.xlim(-5, 15)
# pl.ylim(-1,1.5)
# pl.show()
```

```python
predictors = ['sdss_Mu', 'sdss_Mg', 'sdss_Mr', 'sdss_Mi', 'sdss_Mz']
data = df[predictors]
data = StandardScaler().fit_transform(data)

model = PCA(n_components=2)
projection = model.fit_transform(data)

pl.scatter(projection[:,0][df['galaxy_class']==0], projection[:,1][df['galaxy_class']==0], label='Blue', color='b', s=0.2)
pl.scatter(projection[:,0][df['galaxy_class']==1], projection[:,1][df['galaxy_class']==1], label='Red', color='r', s=0.2)
pl.legend()
pl.xlim(-5, 15)
pl.ylim(-1,1.5)
pl.show()

fig = pl.figure(figsize=(10,4))
pl.subplot(1,2,1)
pl.scatter(projection[:,0][df['galaxy_class']==0], projection[:,1][df['galaxy_class']==0], label='Blue', color='b', s=0.2)
pl.xlim(-5, 15)
pl.ylim(-1,1.5)
pl.subplot(1,2,2)
pl.scatter(projection[:,0][df['galaxy_class']==1], projection[:,1][df['galaxy_class']==1], label='Red', color='r', s=0.2)
pl.xlim(-5, 15)
pl.ylim(-1,1.5)
pl.show()
```
